```python
import os
from dotenv import load_dotenv
import google.generativeai as genai

load_dotenv()
GEMINI_API_KEY = os.getenv('GEMINI_API_KEY')

genai.configure(api_key=GEMINI_API_KEY)

generation_config = genai.GenerationConfig(
    max_output_tokens=500,
    temperature=1.0,
)

model = genai.GenerativeModel(model_name="gemini-1.5-flash")
```

```python
class MyAgent:
    def __init__(self, role : str):
        self.role = role
        self.memory = role

    def listen(self, new_info : str):
        self.memory += new_info
    def talk(self):
        response = model.generate_content(self.memory,
                               generation_config=generation_config)
        return response.text
```

```python
detetive = MyAgent("""
You are Private Detective John Jackson. 
You live in New York, 1929. The great depression has just arrived, and, with that, business has decreased. 
You are worried about how you are going to pay the next rent. Under the dim light of your office, you read the news.
""")

pesquisador = MyAgent("""
You are a research assistant. You read papers and summarize them. You will receive a text 
containing a scientific paper and will summarize it, highlighting its main claims and the evidence supporting each claim. 
Also, you will find out why each claim is relevant to the paper's target audience.
""")

personagem = MyAgent("""
You are Mrs. Jannette Kluperskien Lappone. 
Your father immigrated from Austria to New York when you were 15, and you promptly married a rich entrepreneur from Boston called William Lappone. 
William had kindly supported your family, but with the financial crisis he lost everything and was found dead in his office. 
The police wants to rule this as suicide, which would prevent you from getting your late husband's life insurance. 
You want to hire a private detective to find out who murdered your husband.
""")
```

```python
for _ in range(5):
    text = personagem.talk()
    print("================================= PERSONAGEM ==================================")
    print(text)
    print("===============================================================================")
    detetive.listen(text)
    text2 = detetive.talk()
    print("================================= DETETIVE ==================================")
    print(text2)
    print("===============================================================================")
    personagem.listen(text2)
```
